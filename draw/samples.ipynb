```python
import matplotlib.pyplot as plt
import torch
from matplotlib.ticker import AutoMinorLocator, MaxNLocator

def plot_statistics(paths):
    labels = ['Clean for Train', 'Noise for Train', 'Delete Clean', 'Delete Noise']
    num_plots = len(paths)
    
    # 初始化最大最小值变量，用于后续确定刻度范围
    max_val = float('-inf')
    min_val = float('inf')
    
    # 首先遍历一次路径来更新最大最小值
    for sample_matrix_path in paths:
        sample_matrix = torch.load(sample_matrix_path)
        max_val = max(max_val, sample_matrix.max())
        min_val = min(min_val, sample_matrix.min())
    
    plt.figure(figsize=(20, 4))
    
    for idx, sample_matrix_path in enumerate(paths):
        sample_matrix = torch.load(sample_matrix_path)
        epochs_range = range(1, sample_matrix.shape[0] + 1)
        
        ax = plt.subplot(1, num_plots, idx + 1)
        
        for i, label in enumerate(labels):
            plt.plot(epochs_range, sample_matrix[:, i], label=label)
        
        plt.title(f'Training Statistics Over Epochs\n{sample_matrix_path}')
        plt.xlabel('Epoch')
        plt.ylabel('Number of Samples')
        if idx == 0:
            plt.legend()
        
        # 设置y轴的主要刻度定位器，确保所有图共享相同的刻度范围和间隔
        ax.yaxis.set_major_locator(MaxNLocator(nbins='auto'))
        ax.set_ylim(min_val, max_val+2000)  # 设置统一的y轴范围
        
        # 添加次要刻度
        ax.yaxis.set_minor_locator(AutoMinorLocator())
        plt.tick_params(which='both', width=1)
        plt.tick_params(which='minor', length=4, color='r')
    
    plt.tight_layout()
    plt.show()

# 假设这是您的路径
paths = [

    '../sta/sample_matrix_sym_0.6.pt',
    '../sta/sample_matrix_sym_0.6 2c.pt'
]

plot_statistics(paths)
```
